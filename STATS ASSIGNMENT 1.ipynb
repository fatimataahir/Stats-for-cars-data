```python
# Specify the numerical columns for which you want to calculate variance and standard deviation
numerical_columns = ['Sales_in_thousands', '__year_resale_value', 
                     'Price_in_thousands', 'Engine_size', 'Horsepower']

# Calculate variance and standard deviation
variances = df[numerical_columns].var()
std_devs = df[numerical_columns].std()

# Display the results
for column in numerical_columns:
    print(f"Variance of {column}: {variances[column]:.2f}")
    print(f"Standard Deviation of {column}: {std_devs[column]:.2f}")
    print("----------------------")
```

```python
# Specify the numerical columns for which you want to calculate IQR and identify outliers
numerical_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower']

# Identify outliers using the IQR method
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Identify outliers
    outliers = df[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)]
    
    # Display results
    print(f"Column: {column}")
    print(f"IQR: {IQR:.2f}")
    print(f"Q1: {Q1:.2f}")
    print(f"Q3: {Q3:.2f}")
    print(f"Number of outliers: {len(outliers)}")
    print("----------------------")
```
